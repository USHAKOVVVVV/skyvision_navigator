```python
from ultralytics import YOLO

# –ó–∞–≥—Ä—É–∑–∫–∞ –ø—Ä–µ–¥–æ–±—É—á–µ–Ω–Ω–æ–π –º–æ–¥–µ–ª–∏
model = YOLO('–ú–æ–¥–µ–ª—é—Å–∏/yolov8n_gpu_simple_1.pt')  # –¥–ª—è —Å–µ–≥–º–µ–Ω—Ç–∞—Ü–∏–∏
```

```python
# –ù–∞–π—Ç–∏ –∏ –≤—ã–≤–µ—Å—Ç–∏ –¢–û–õ–¨–ö–û –ø—Ä–æ–±–ª–µ–º–Ω—ã–µ —Ñ–∞–π–ª—ã
import os

def find_only_problem_files():
    print("üéØ –ü–æ–∏—Å–∫ –¢–û–õ–¨–ö–û –ø—Ä–æ–±–ª–µ–º–Ω—ã—Ö —Ñ–∞–π–ª–æ–≤...")
    
    problem_files = []
    
    # –ü–∞–ø–∫–∏ –¥–ª—è –ø—Ä–æ–≤–µ—Ä–∫–∏
    label_folders = [
        'train/labels',
        'valid/labels', 
        'test/labels'
    ]
    
    for labels_dir in label_folders:
        if not os.path.exists(labels_dir):
            continue
            
        files = [f for f in os.listdir(labels_dir) if f.endswith('.txt')]
        
        for file in files:
            file_path = os.path.join(labels_dir, file)
            
            with open(file_path, 'r') as f:
                lines = f.readlines()
            
            if not lines:
                continue
                
            has_detection = False
            has_segmentation = False
            
            for line in lines:
                parts = line.strip().split()
                if len(parts) >= 5:
                    if len(parts) == 5:
                        has_detection = True
                    else:
                        has_segmentation = True
            
            # –î–æ–±–∞–≤–ª—è–µ–º —Ç–æ–ª—å–∫–æ –ø—Ä–æ–±–ª–µ–º–Ω—ã–µ —Ñ–∞–π–ª—ã
            if has_detection:
                if has_segmentation:
                    problem_files.append(('üîÑ –°–ú–ï–®–ê–ù–ù–´–ô', file_path))
                else:
                    problem_files.append(('‚ùå –î–ï–¢–ï–ö–¶–ò–Ø', file_path))
    
    # –í—ã–≤–æ–¥–∏–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã
    if problem_files:
        print(f"\nüö® –ù–∞–π–¥–µ–Ω–æ {len(problem_files)} –ø—Ä–æ–±–ª–µ–º–Ω—ã—Ö —Ñ–∞–π–ª–æ–≤:")
        print("-" * 80)
        
        for problem_type, file_path in problem_files:
            print(f"{problem_type}: {file_path}")
            
            # –ü–æ–∫–∞–∑—ã–≤–∞–µ–º —Å–æ–¥–µ—Ä–∂–∏–º–æ–µ –ø—Ä–æ–±–ª–µ–º–Ω–æ–≥–æ —Ñ–∞–π–ª–∞
            try:
                with open(file_path, 'r') as f:
                    content = f.read().strip()
                lines = content.split('\n')
                print(f"   –°–æ–¥–µ—Ä–∂–∏–º–æ–µ ({len(lines)} —Å—Ç—Ä–æ–∫):")
                for i, line in enumerate(lines[:3]):  # –ø–µ—Ä–≤—ã–µ 3 —Å—Ç—Ä–æ–∫–∏
                    parts = line.split()
                    print(f"   –°—Ç—Ä–æ–∫–∞ {i}: {len(parts)} –∑–Ω–∞—á–µ–Ω–∏–π - {parts}")
                if len(lines) > 3:
                    print(f"   ... –∏ –µ—â–µ {len(lines) - 3} —Å—Ç—Ä–æ–∫")
                print()
            except Exception as e:
                print(f"   –û—à–∏–±–∫–∞ —á—Ç–µ–Ω–∏—è: {e}")
                print()
        
        print("-" * 80)
    else:
        print("‚úÖ –ü—Ä–æ–±–ª–µ–º–Ω—ã—Ö —Ñ–∞–π–ª–æ–≤ –Ω–µ –Ω–∞–π–¥–µ–Ω–æ!")
    
    return problem_files

# –ó–∞–ø—É—Å–∫–∞–µ–º –ø–æ–∏—Å–∫ –ø—Ä–æ–±–ª–µ–º–Ω—ã—Ö —Ñ–∞–π–ª–æ–≤
problem_files = find_only_problem_files()
```

```python
results = model.train(
    data='data.yaml',
    epochs=20,          # –≠–ø–æ—Ö–∏
    
    imgsz=640, # –†–∞–∑–º–µ—Ä –∏–∑–æ–±—Ä–∞–∂–µ–Ω—å–∫–∏ —Å—Ç–∞–≤–∏–º 640, —Ç.–∫. –≤ —ç—Ç–æ–º —Ñ–æ—Ä–º–∞—Ç–µ —Ä–∞–∑–º–µ—á–∞–µ–º
    
    batch=2,            # –ü–æ —Å–∫–æ–∫ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–µ–∫ –±–µ—Ä–µ–º –∑–∞ –ø—Ä–æ–≥–æ–Ω
    
    patience=15,    #–¢–µ—Ä–ø–µ–Ω–∏–µ, —Ö–∑ —á–µ —ç—Ç–æ –∑–Ω–∞—á–∏—Ç ?

    device='cuda',      # –ò—Å–ø–æ–ª—å–∑—É–µ–º CPU, –µ—Å–ª–∏ –Ω–µ—Ç –≤–∏–¥–µ–æ–∫–∞—Ä—Ç—ã, –µ—Å–ª–∏ –µ—Å—Ç—å, —Ç–æ —Å—Ç–∞–≤–∏–º 'cuda'

    workers=2,          # –ú–∏–Ω–∏–º—É–º workers
    
    lr0=0.005,   # –õ–µ—Ä–Ω–∏–Ω–≥ —Ä—ç–π—Ç ?

    optimizer='AdamW',   # Adam - –ª–µ–≥–∫–∏–π –æ–ø—Ç–∏–º–∏–∑–∞—Ç–æ—Ä, AdamW - –ñ–µ—Å—Ç—á–µ

    # weight_decay = 0.0005,   # –†–µ–≥—É–ª—è—Ä–∏–∑–∞—Ü–∏—è ?
    
    name='yolov8n_gpu_updgrade_1',
    
    exist_ok=True
)
```

```python
from ultralytics import YOLO
import os

# –ü—Ä–æ–≤–µ—Ä–∫–∞ —Å—Ç—Ä—É–∫—Ç—É—Ä—ã –¥–∞—Ç–∞—Å–µ—Ç–∞
def check_dataset_structure(dataset_path):
    print("–ü—Ä–æ–≤–µ—Ä–∫–∞ —Å—Ç—Ä—É–∫—Ç—É—Ä—ã –¥–∞—Ç–∞—Å–µ—Ç–∞...")
    
    required_dirs = [
        'train/images',
        'valid/images', 
        'train/labels',
        'valid/labels'
    ]
    
    for dir_path in required_dirs:
        full_path = os.path.join(dataset_path, dir_path)
        if not os.path.exists(full_path):
            print(f"‚ùå –û—Ç—Å—É—Ç—Å—Ç–≤—É–µ—Ç: {full_path}")
        else:
            print(f"‚úÖ –ù–∞–π–¥–µ–Ω–æ: {full_path}")
            
    # –ü—Ä–æ–≤–µ—Ä–∫–∞ —Ñ–∞–π–ª–æ–≤ labels
    labels_train_path = os.path.join(dataset_path, 'train/labels')
    if os.path.exists(labels_train_path):
        label_files = os.listdir(labels_train_path)
        if label_files:
            sample_label = os.path.join(labels_train_path, label_files[0])
            with open(sample_label, 'r') as f:
                content = f.read().strip()
                print(f"–ü—Ä–∏–º–µ—Ä —Ñ–∞–π–ª–∞ —Ä–∞–∑–º–µ—Ç–∫–∏: {content}")
                
                # –ü—Ä–æ–≤–µ—Ä–∫–∞ —Ñ–æ—Ä–º–∞—Ç–∞
                if content:
                    parts = content.split()
                    if len(parts) > 5:
                        print("‚úÖ –§–æ—Ä–º–∞—Ç: –°–µ–≥–º–µ–Ω—Ç–∞—Ü–∏—è (–ø–æ–ª–∏–≥–æ–Ω—ã)")
                    else:
                        print("‚ùå –§–æ—Ä–º–∞—Ç: –î–µ—Ç–µ–∫—Ü–∏—è (bbox), –∞ –Ω—É–∂–Ω–∞ —Å–µ–≥–º–µ–Ω—Ç–∞—Ü–∏—è!")

# –ü—Ä–æ–≤–µ—Ä–∫–∞
dataset_path = ""  # —É–∫–∞–∂–∏—Ç–µ –≤–∞—à –ø—É—Ç—å
check_dataset_structure(dataset_path)
```

```python
model_path = 'runs/segment/yolov8n_gpu_upgrade_1/weights/best.pt'
model = YOLO(model_path)
```

```python
# –í–∞–ª–∏–¥–∞—Ü–∏—è –Ω–∞ —Ç–µ—Å—Ç–æ–≤—ã—Ö –¥–∞–Ω–Ω—ã—Ö
results = model.val(
    data='data.yaml',
    split='test',    # –∏—Å–ø–æ–ª—å–∑—É–µ–º –¢–ï–°–¢–û–í–´–ï –¥–∞–Ω–Ω—ã–µ
    conf=0.25, # —É—Ä–æ–≤–µ–Ω—å –¥–æ—Å—Ç–æ–≤–µ—Ä–Ω–æ—Å—Ç–∏
    iou=0.3, #–º–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–π –ø–æ—Ä–æ–≥ –ø–µ—Ä–µ—Å–µ—á–µ–Ω–∏—è –æ–±—ä–µ–∫—Ç–æ–≤
    device='cpu'
)

print(f"üìä –†–µ–∑—É–ª—å—Ç–∞—Ç—ã –æ—Ü–µ–Ω–∫–∏:")
print(f"mAP50: {results.box.map50:.3f}")

# print(results.box)

# –ï—Å–ª–∏ mAP50 < 0.6 ‚Üí –Ω—É–∂–Ω–æ —É–ª—É—á—à–∞—Ç—å –º–æ–¥–µ–ª—å
# –ï—Å–ª–∏ mAP50 > 0.8 ‚Üí –æ—Ç–ª–∏—á–Ω—ã–π —Ä–µ–∑—É–ª—å—Ç–∞—Ç!
```
